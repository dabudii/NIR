```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")
```

```python
model.train(
    data="/content/drive/MyDrive/mollusca_dataset/data.yaml",
    epochs=100,
    patience = 10,
    batch=16,
    imgsz=640,
    save=True,
    save_period=5,
    cache="disk",
    device=0,
    workers=8,
    project="/content/drive/MyDrive/NIR_proj/results",
    name="NIR_proj",
    optimizer="auto",
    close_mosaic = 10,
    dropout = 0.2,
    weight_decay=0.05,
    degrees=45.0,
    fraction=1.0,
    profile=False,
    lr0=0.01,
    lrf=0.01,
    box=7.5,
    cls=0.5,
    dfl=2.0,
    overlap_mask=True,
    val = True,
    exist_ok=False,
    pretrained=False,
    amp = True,
    plots = True,
    cos_lr = True,
    rect = True,
    multi_scale = True
)
```

```python
best_model = YOLO("/content/drive/MyDrive/NIR_proj/results/NIR_proj5/weights/best.pt")

metrics = best_model.val(
    data="/content/drive/MyDrive/mollusca_dataset/data.yaml",
    batch=16,
    imgsz=640,
    conf=0.25,
    iou=0.5,
    name="mollusca_eval"
)

print(f"mAP50-95: {metrics.box.map}")
print(f"mAP50: {metrics.box.map50}")
print(f"mAP75: {metrics.box.map75}")
print(f"Precision: {metrics.box.mp}")
print(f"Recall: {metrics.box.mr}")
```
